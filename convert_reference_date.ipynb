```python
df = pd.read_csv ('Working_prediction.csv')

print(df.columns)

df
```

```python
df['DT'] = 0
df['weekday'] = 0 

for i in range (9494):
    
    temp = df.iloc[i]
    
    df.iloc[i, 29] = dt.datetime(int(temp[7]), int(temp[8]), int(temp[9]))
    
    df.iloc[i, 30] = (dt.datetime(int(temp[7]), int(temp[8]), int(temp[9]))).weekday()
    
df
    
```

```python
df

df['ref_date'] = 0 

for i in range (9494):
    
    temp = df.iloc[i] 
    
    if (temp[30] == 0):
        
        if temp[1] > 9 and temp[10] > 30:
            
            df.iloc[i,31] = temp[29] + dt.timedelta(days=1)
            
        else:
            df.iloc[i,31] = temp[29] 
            
    elif (temp[30] == 5) or (temp[30] == 6):
        
        change = 7 - temp[30]
        
        df.iloc[i,31] = temp[29] + dt.timedelta(days=int(change))
    
    else: # 1234, Tue, Wed, Thur, or Fri
    
        if temp[1] > 9 and temp[10] > 30:
            
            if (temp[30] == 4):
                df.iloc[i,31] = temp[29] + dt.timedelta(days=3)
            else:
                df.iloc[i,31] = temp[29] + dt.timedelta(days=1)
            
        else:
            df.iloc[i,31] = temp[29] 

```
